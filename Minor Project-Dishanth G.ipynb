## Creating a new Column called "grades" where we give grades for the student based on their percentage:

```python
def determine_grade(scores):
    if scores >= 90 and scores <= 100:
        return 'S'
    elif scores >= 80 and scores < 90:
        return 'A'
    elif scores >= 70 and scores < 80:
        return 'B'
    elif scores >= 60 and scores < 70:
        return 'C'
    elif scores >= 55 and scores < 60:
        return 'D'
    elif scores >=50 and scores <55:
        return 'E'
    else:
        return 'F'
    
st['grades']=st['percentage'].apply(determine_grade)
```

### Inference:
- Here we have created a function which takes the percentage as an argument and returns a grade.
- Below is the grading criteria:
    <ul>
    <li>Percentage between 90 and 100 = S GRADE</li>
    <li>Percentage between 80 and 90 = A GRADE</li>
    <li>Percentage between 70 and 80 = B GRADE</li>
    <li>Percentage between 60 and 70 = C GRADE</li>
    <li>Percentage between 55 and 66 = D GRADE</li>
    <li>Percentage between 50 and 55 = E GRADE</li>
    <li>Percentage below 50 = F GRADE</li>
    </ul>
- S grade is considered superior followed by A,B,C,D,E,F.
